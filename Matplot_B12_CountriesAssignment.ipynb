```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('countries.csv')

df
```

```python
aus = df[df.country == 'Australia']

aus
```

```python
bah = df[df.country == 'Bahrain']

bah
```

```python
plt.plot(aus.population,aus.year)

plt.plot(bah.population,bah.year)

plt.title('Comparision of Australia and Bahrain population')
plt.xlabel('population')

plt.ylabel('Year')

plt.legend(['Indicates Australia Info','Indicates Bahrain info']) # it will be always accepting the input as list of args

plt.show()
```
